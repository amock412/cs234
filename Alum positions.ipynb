In this notebook, the aim is to analyze the positions that Wellesley alums hold at their current workplace. We used data collected from the Hive and LinkedIn to generate the word clouds for each cohort of alums.

```python
import os
import time

import numpy as np
import pandas as pd
import csv
```

Read in CSV file into Pandas dataframe with all the data on Wellesley alums from LinkedIn and Hive. 

```python
allAlumDf = pd.read_csv('all_alum_data.csv')
#allAlumDf.head()
```

Create separate dataframes for each of each of the Wellesley alums cohorts based on their grad year. The cohorts are: 2015-2020, 2010 - 2014, 2000-2009, 1990-1999, 1980-1989, 1970-1979.

```python
grad2015 = allAlumDf[allAlumDf['grad_year'].isin([2015, 2016, 2017, 2018, 2019, 2020])]
grad2010 = allAlumDf[allAlumDf['grad_year'].isin([2014, 2013, 2012, 2011, 2010])]
grad2000 = allAlumDf[allAlumDf['grad_year'].isin([2009, 2008, 2007, 2006, 2005, 2004, 2003, 2002, 2001, 2000])]
grad1990 = allAlumDf[allAlumDf['grad_year'].isin([1999, 1998, 1997, 1996, 1995, 1994, 1993, 1992, 1991, 1990])]
grad1980 = allAlumDf[allAlumDf['grad_year'].isin([1989, 1988, 1987, 1986, 1985, 1984, 1983, 1982, 1981, 1980])]
grad1970 = allAlumDf[allAlumDf['grad_year'].isin([1979, 1978, 1977, 1976, 1975, 1974, 1973, 1972, 1971, 1970])]
```

Drop all the alums who do not have a title.

```python
grad2015clean = grad2015['title'].dropna()
grad2010clean = grad2010['title'].dropna()
grad2000clean = grad2000['title'].dropna()
grad1990clean = grad1990['title'].dropna()
grad1980clean = grad1980['title'].dropna()
grad1970clean = grad1970['title'].dropna()
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt 
```

Create word cloud for 2015-2020 cohort.

```python
wordcloud2015 = WordCloud().generate("".join(grad2015clean))
plt.imshow(wordcloud2015, interpolation = 'bilinear')
plt.axis("off")
plt.show()
wordcloud2015.to_file("wordcloud2015.png")
```
