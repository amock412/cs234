```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read CSV file with all alums into dataframe.

```python
allAlumDf = pd.read_csv('all_alum_data.csv')
#allAlumDf.head()
```

Save count of unique companies into a dictionary. Remove certain "companies" such as 'self-employed' or 'Wellesley College' because they are not legitimate company names. Wellesley College was removed, but not other schools, as there is a possbility that alums went to other schools to take on research positions. Amazon Web Services was removed because it overlaps with Amazon.

```python
allAlumDf['current_employer'].value_counts()
counts = allAlumDf['current_employer'].value_counts().to_dict()
```

```python
counts.pop('Self-Employed', None)
counts.pop('unemployed', None)
counts.pop('No companies provided', None)
counts.pop('freelance', None)
counts.pop('Self-employed', None)
counts.pop('Wellesley College', None)
counts.pop("Freelance", None)
counts.pop("Self Employed", None)
counts.pop('Amazon Web Services (AWS)', None)
```

Create bar chart that shows the frequency of alums at the top 50 companies. 

```python
company = list(counts.keys())
values = list(counts.values())

fig = plt.figure(figsize = (20, 10))

plt.barh(company[:50], values[:50])
plt.xlabel('Frequency')
plt.ylabel('Company')
plt.title("Bar Plot of Top 50 Most Popular Companies")
plt.savefig("companybarplot.png")
```
